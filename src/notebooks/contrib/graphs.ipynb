```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def function(x, eps, sigma):
    dpot = (sigma/x)**6
    y = 4.0*eps*(dpot-1.0)*dpot
    return y
```

```python
x = np.arange(3.0, 10.0, 0.1)
```

```python
eps = 0.2385
sigma = 3.4
plt.plot(x, function(x, eps, sigma))
plt.xlim(2, 10)
plt.ylim(-0.4, 0.5)
plt.xlabel(r'r ($\AA$)')
plt.ylabel(r'U(r) (kcal/mol)')
plt.text(6, 0.2, r'$U(r) = 4\varepsilon\left[\left(\frac{\sigma}{r}\right)^{12}-\left(\frac{\sigma}{r}\right)^{6}\right]$', fontsize = 15)
plt.axhline(y=0, color='black', linestyle='--', label='Horizontal Line')
plt.axvline(x=sigma, color='black', linestyle='--', label='Horizontal Line')
plt.annotate(r'$r = \sigma$', xy = (sigma, 0), xytext = (4, 0.1), arrowprops=dict(arrowstyle="->"))
plt.annotate(r'$(2^{1/6}\sigma, \varepsilon)$', xy = (sigma*2**(1/6), -eps), xytext = (4.5, -0.3), arrowprops=dict(arrowstyle="->"))
plt.figure().set_figwidth(10)
plt.savefig('lennard-jones.jpg')
```
