```python
from m3l import structure as strc
from m3l.dr_module import forcefield as nl
from m3l import force_field as ff
import numpy as np
```

```python
sistema = strc.System(15, 15, 15, 'h2.xyz')
print(sistema)
```

```python
r1 = [u['x'] for u in sistema.atoms]
r1 = np.array(r1)
r2 = [u['y'] for u in sistema.atoms]
r2 = np.array(r2)
r3 = [u['z'] for u in sistema.atoms]
r3 = np.array(r2)
print(r1, r2, r3)
```

```python
forcefield = ff.ForceField(1.0)
rcutoff = forcefield.rcutoff
print(rcutoff)
```

```python
nl = nl.neighbour_list(r1, r2, r3, rcutoff, sistema.getNatoms())
nl
```
