```python
from m3l.structure import System as sys
import json
#
system = sys()
system.loadSystem('system_new.json')
system.atoms
```

```python
x = [item[1].item() for item in system.atoms]
y = [item[2].item() for item in system.atoms]
z = [item[3].item() for item in system.atoms]
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z)
ax.view_init(90, -90, 0)
#ax.axis('off')
```
