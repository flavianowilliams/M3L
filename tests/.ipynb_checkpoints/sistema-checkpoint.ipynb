# Teste da preparação do sistema físico

```python
from m3l import structure as strc
```

```python
sistema = strc.System(25,25,25,'h2.xyz')
print(sistema)
```

```python
print(sistema.atoms)
```

```python
with open('h2.xyz') as xyz_file:
    natom = xyz_file.readline()
    natom = int(natom)
    xyz_file.readline()
    atoms = list()
    for i in range(natom):
        p1, p2, p3, p4 = xyz_file.readline().split(' ', maxsplit=3)
        p2 = float(p2)
        p3 = float(p3)
        p4 = float(p4)
        atoms.append({'id': i, 'atom': p1, 'x': p2, 'y': p3, 'z': p4})
        
print(atoms)
```
